```python
import pandas as pd
```

```python
NSL=pd.read_csv("NSL_kdd.csv")
# print(NSL.to_string())
NSL
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
labelprotocol = le.fit_transform(NSL["protocol_type"])
labelservice = le.fit_transform(NSL["service"])
labelflag = le.fit_transform(NSL["flag"])
labelclass = le.fit_transform(NSL["class"])
```

```python
Data = NSL.drop(["protocol_type","class","flag","service"], axis='columns')
```
